```
import itertools as itt
import matplotlib.pyplot as plt
from time import time
```

**Disclaimer:** We used list data structure from python because it has an append time complexity of $O(1)$ so it won't heavily interfere with this process.

```
#Recursive Fibonacci
#Running until 45 due to the delay when navigating the tree generated for this recursion
def RFibonacci(n): 
  if n <= 1:
    return n
  else:
    return (RFibonacci(n-1) + RFibonacci(n-2))
      
nRecF = 46    
```

```
tR = []
for t in range(nRecF):
  ti = time()
  RFibonacci(t)
  tf = time()
  tR.append(tf-ti)
```

```
XRaxis = []
for i in range(nRecF):
  XRaxis.append(i)


plt.plot(XRaxis,tR,label = 'Tiempo De Fibonacci Recursivo')
plt.plot(XRaxis,XRaxis,label = 'Tiempo De Ejecucion Lineal')
plt.title("Comparacion en tiempos de ejecucion")
plt.xlabel('Numero de elementos')
plt.ylabel('Tiempo de ejecucion')
plt.legend()
plt.show()
```

**Analysis:** Recursion turns out to be a problem when doing calculations like this, what can be understood behind this implementation is that every value that its estimated generates a tree behind 

For a simple calculation using the three structure we have to navigate through $2^n$ nodes, where the root is the objective and the leafs from the bottom are the initial steps of this recursion, hence the time this requires is:

$$ T_{n} = 2^n = O(2^n) $$

therefore that explains why the behaviour within the time grows at an exponential rate.



```
#Semi recursive Fibonacci using dictionaries
def SRFibonacci(n, computed = {0: 0, 1: 1}):
  if n not in computed:
    computed[n] = SRFibonacci(n-1, computed) + SRFibonacci(n-2, computed)
  return computed[n]

nSRecF = 51  
```

```
tSR = []
for t in range(nSRecF):
  ti = time()
  SRFibonacci(t)
  tf = time()
  tSR.append(tf-ti)
```

```
XSRaxis = []
for i in range(nSRecF):
  XSRaxis.append(i)



plt.plot(XSRaxis,tSR,label = 'Tiempo De Fibonacci Recursivo')
plt.plot(XSRaxis,XSRaxis,label = 'Tiempo De Ejecucion Lineal')
plt.title("Comparacion en tiempos de ejecucion")
plt.xlabel('Numero de elementos')
plt.ylabel('Tiempo de ejecucion')
plt.legend()
plt.show()
```

```
plt.plot(XSRaxis,tSR,label = 'Tiempo De Fibonacci Recursivo')
plt.title("Comparacion en tiempos de ejecucion")
plt.xlabel('Numero de elementos')
plt.ylabel('Tiempo de ejecucion')
plt.legend()
plt.show()
```

```
#Iterative Fibonacci
def IFibonacci(n): 
  a, b = 0, 1
  for i in range(0, n):
    a, b = b, a + b
  return a
      
nIterF = 501  
```

```
tI = []
for t in range(nIterF):
  ti = time()
  IFibonacci(t)
  tf = time()
  tI.append(tf-ti)
```

```
XIaxis = []
for i in range(nIterF):
  XIaxis.append(i)


plt.plot(XIaxis,tI,label = 'Tiempo De Fibonacci Iterativo')
plt.plot(XIaxis,XIaxis,label = 'Tiempo De Ejecucion Lineal')
plt.title("Comparacion en tiempos de ejecucion")
plt.xlabel('Numero de elementos')
plt.ylabel('Tiempo de ejecucion')
plt.legend()
plt.show()
```

```
plt.plot(XIaxis,tI,label = 'Tiempo De Fibonacci Iterativo')
plt.title("Comparacion en tiempos de ejecucion")
plt.xlabel('Numero de elementos')
plt.ylabel('Tiempo de ejecucion')
plt.legend()
plt.show()
```

**Analysis:** This two methods (with the semi-recursive) behave better because of the number of operations that have to be done in order to obtain a number, when we saveour numbers we are automatically reducing time complexity, we reduce the use of space therefore the time complexity is:

$$ T_n = O(n) = n $$

When Comparing both grafics we can see that when considering only the time in iterative Fibonaci, it follows a linear tendency with small peaks. Yet it is completely efficent and can be used to do this calculations fastly without losing information or data
