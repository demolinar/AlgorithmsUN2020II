```
import itertools as itt
import matplotlib.pyplot as plt
```

```
v = []
def bubbleSort(v):
  pasos = 0 #initial steps
  n = len(v) #amount of elements to sort
  flag = True
  while flag==True and n>1:
    flag = False
    for i in range(len(v)-1):
      if v[i]>v[i+1]: # if the next previous element is smaller than the next one
        v[i],v[i+1]=v[i+1],v[i] #Swap values between them
        pasos = pasos + 1 #and add another step
        flag = True #mark them as swapped
    n -= 1 
  return pasos
```

```
#permutations tuples generated with itertols
perm = itt.permutations([1,2,3]) 
dist = []
for i in list(perm):
  dist.append(bubbleSort(list(i))) #Storing the values obtained with the steps for sorting

#plotting
plt.title('Distribucion del Bubble Sort')
plt.hist(dist, max(dist)+1, alpha=1, edgecolor = 'black',  linewidth=1)
plt.xlabel('Numero de pasos')
plt.ylabel('Frecuencia de pasos de ordenamiento')
plt.show()
```

**Conclusions:** When plotting the frequency of steps taken for a list to be sorted by Bubble sort based on its level of disorder, we can conclude that on average the distribution behave like a normal distribution centered almost arround at.

$$2 \times len(v)$$

Where $len(v)$ is the length of the list passed to this method,  we can observe that the number of different values is nearly $4 \times len(v)$, yet no matter the length of the list it has this same behaviour.

The problem beggins when calculating the permutions as is known a permutation is a way to distribute $N$ elements in $n$ spaces , when $N=n$ then the permutations are $P_{N} = N!$ which is almost an exponential problem, and turns out to be hard to compute

Now talking about **Bubble Sort**, whit a total of $\frac{n(n+1)}{2}$ operations, time complexity for Bubble sort is:


*    Worst Case Time Complexity $O(n)=O(n^2)$
*    Best Case Time Complexity $\Omega(n)= O(n)$
*    Average Time Complexity $\Theta(n)=O(n^2)$
*    Space Complexity: $O(1)$
